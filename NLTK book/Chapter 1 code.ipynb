#### 3.4 Counting Other Things
Distributions of word lengths

````python
LenList1 = [len(i) for i in text1]
FDLen1 = FreqDist(LenList1)
print(FDLen1)
print(FDLen1.most_common(10))
print(FDLen1.max())
print(FDLen1[3])
print(FDLen1.freq(3))
````

#### 28.  Define a function `percent(word, text)` that calculates how often a given word occurs in a text, and expresses the result as a percentage.
