```python
import nltk
from nltk.corpus import reuters
from nltk.corpus import brown
```

```python
fileids = nltk.corpus.reuters.fileids()
samples = ['memorandum','memo','gym','gymnasium','gas','gasoline','exam','examination']
categories = nltk.corpus.reuters.categories()
cfdReuter = nltk.ConditionalFreqDist(
    (cate,word)
    for cate in categories
    for word in reuters.words(categories=cate)
)
cfdReuter.tabulate(samples = samples)
cfdReuter.plot(samples = samples)
```

```python
categories = nltk.corpus.brown.categories()
cfdBrown = nltk.ConditionalFreqDist(
    (cate,word)
    for cate in categories
    for word in brown.words(categories=cate)
)
cfdBrown.tabulate(samples = samples)
```
