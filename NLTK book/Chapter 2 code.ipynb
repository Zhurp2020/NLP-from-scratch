```python
def UnusualWords(text):
    TextVocab = set(i.lower() for i in text if i.isalpha())
    EngVocab = set(i.lower() for i in nltk.corpus.words.words())
    return list(EngVocab-TextVocab)
print(UnusualWords(nltk.corpus.gutenberg.words('austen-sense.txt'))[:20])
```

#### 4.2 A Pronouncing Dictionary

```python
from nltk.corpus import cmudict
for word,pron in cmudict.entries():
    if len(pron) == 3:
        pr1,pr2,pr3 = pron
        if pr1 == 'P' and pr3 == 'T' :
            print(word,pr1,pr2,pr3,end=' ')
```

```python
def stress(pron):
    stress = [i for phone in pron for i in phone  if i.isdigit()]
    return stress
print([i for i,pron in cmudict.entries() if stress(pron) == ['0', '1', '0', '2', '0']][:10])
```

```python
pwords = [(pron[0]+'-'+pron[-1],word) for word,pron in cmudict.entries() if len(pron) ==3 and pron[0] == 'P']
CFD5 = nltk.ConditionalFreqDist(pwords)
for condition in CFD5.conditions():
    #print(list(CFD5[condition]))
    if len(CFD5[condition]) >= 10 :
        print(condition,' '.join(CFD5[condition])) 
```

#### 14. Define a function `supergloss(s)` that takes a synset `s` as its argument and returns a string consisting of the concatenation of the definition of `s`, and the definitions of all the hypernyms and hyponyms of `s`.

```python
def supergloss(s):
    def_ = s.definition()
    hypernyms = '\n'.join([i.definition() for i in s.hypernyms()])
    hyponyms = '\n'.join([i.definition() for i in s.hyponyms()])
    return def_ + '\n' + hypernyms + '\n' + hyponyms
print(supergloss(wn.synset('car.n.01')))
```

#### 24. Define a function `hedge(text)` which processes a text and produces a new version with the word 'like' between every third word.

```python
def hedge(text):
    NewText = []
    for i in range(len(text)):
        NewText.append(text[i])
        if (i+1) % 3 == 0 :
            NewText.append('like')
    return NewText
print(hedge(['a','b','c','d','e']))
```

#### 25. Define a function `find_language()` that takes a string as its argument, and returns a list of languages that have that string as a word. Use the `udhr` corpus and limit your searches to files in the Latin-1 encoding.

```python
from nltk.corpus import udhr
def find_language(word):
    result = ''
    fileids = [i for i in udhr.fileids() if 'Latin1' in i]
    for fid in fileids:
        if word in udhr.words(fid):
            result += ' '.join(fid.split('-')[:-1]) 
            result += '/'
    return result
find_language('da')
```

#### 28. Use one of the predefined similarity measures to score the similarity of each of the following pairs of words. Rank the pairs in order of decreasing similarity. How close is your ranking to the order given here, an order that was established experimentally by (Miller & Charles, 1998): car-automobile, gem-jewel, journey-voyage, boy-lad, coast-shore, asylum-madhouse, magician-wizard, midday-noon, furnace-stove, food-fruit, bird-cock, bird-crane, tool-implement, brother-monk, lad-brother, crane-implement, journey-car, monk-oracle, cemetery-woodland, food-rooster, coast-hill, forest-graveyard, shore-woodland, monk-slave, coast-forest, lad-wizard, chord-smile, glass-magician, rooster-voyage, noon-string.

```python
WordList = ['car-automobile', 'gem-jewel', 'journey-voyage', 'boy-lad', 'coast-shore', 'asylum-madhouse', 'magician-wizard', 'midday-noon', 'furnace-stove', 'food-fruit', 'bird-cock', 'bird-crane', 'tool-implement', 'brother-monk', 'lad-brother', 'crane-implement', 'journey-car', 'monk-oracle', 'cemetery-woodland', 'food-rooster', 'coast-hill', 'forest-graveyard', 'shore-woodland', 'monk-slave', 'coast-forest', 'lad-wizard', 'chord-smile', 'glass-magician', 'rooster-voyage', 'noon-string']
def sim(w1,w2):
    ssw1 = wn.synsets(w1)[0]
    ssw2 = wn.synsets(w2)[0]
    return ssw1.path_similarity(ssw2)
for i in WordList:
    w1,w2 = i.split('-')
    print(i,sim(w1,w2))
```
